## Step 2: HuggingFace Login

```python
from huggingface_hub import login
import os
# Add your HF token to Kaggle Secrets: https://www.kaggle.com/docs/secrets
os.environ["HF_TOKEN"] = "YOUR_HF_TOKEN"
login(token=os.environ["HF_TOKEN"])
```

## Step 3: Run Fine-tuning

```python
import subprocess

result = subprocess.run([
    "python3", "training/train_extended_context.py",
    "--model-path", "Qwen/Qwen2.5-Coder-1.5B",
    "--data-path", "training/training-data/tool_examples_combined.jsonl",
    "--output-dir", "/kaggle/working/stack-2.9-128k",
    "--context-length", "131072",
    "--lora-rank", "64",
    "--epochs", "3",
    "--push-to-hub",
    "--hub-model-id", "YOUR_USERNAME/stack-2.9-128k"
], cwd="/kaggle/working/stack-2.9", env=dict(os.environ, HF_TOKEN=os.environ["HF_TOKEN"]))

print(result.stdout[-3000:] if result.stdout else "No stdout")
print(result.stderr[-1000:] if result.stderr else "No stderr")
```
